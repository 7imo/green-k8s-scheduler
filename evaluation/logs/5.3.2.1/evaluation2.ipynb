```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create data
x=["17:00:00", "17:04:00", "17:08:00", "17:12:00", "17:16:00", "17:20:00", "17:24:00", "17:28:00", "17:32:00", "17:36:00"]
consumption=[151.31,151.82,133.39,2145.83,2141.43,3721.44,3767.22,4617.62,4654.47,4683.73]
excess=[1090.39,1417.08,1435.51,0,1110.37,0,0,0,0,0]

# Basic stacked area chart.
plt.figure(figsize=(9,3))
plt.stackplot(x,consumption, excess, labels=['Consumption','Renewable Excess'], colors = ['lightgrey', 'mediumseagreen'])
plt.plot([], [], '--', color="firebrick", label="Pod Deployment")
plt.legend(loc='upper left')
plt.ylim(0, 7000)
plt.xlabel("Time")
plt.ylabel("Watt")
plt.title("172.20.54.144")
xposition = ["17:08:00", "17:16:00", "17:24:00"]
for xc in xposition:
    plt.axvline(x=xc, color='firebrick', linestyle='--')

plt.show()
```
